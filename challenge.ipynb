```python
#funçao para pegar id baseados em um campo e seu respectivo valor
def getidfromfield(table: str, field: str, value: str):
    try:
        cursor = con.cursor(buffered = True)
        select = f"""SELECT id from {table}
                        where
                        {field} = '{value}'"""
        cursor.execute(select)
        id = cursor.fetchall()
        id = id[0][0]
        con.commit()
    except Exception as e:
        print(f'Erro: {e}')
    finally:
        cursor.close()
    return id
```

```python
#Criando tabelas no banco
script = []  #lista que vai receber os scripts de criação
table_emitente = """CREATE TABLE Emitentes (
    id int AUTO_INCREMENT PRIMARY KEY,
    cnpj varchar(14),
    nome_empresa nvarchar (50),
    nome_fantasia nvarchar (50),
    logradouro nvarchar (100),
    numero nvarchar(8),
    complemento nvarchar (50),
    bairro nvarchar (30),
    cod_municipio varchar(7),
    municipio nvarchar (20),
    estado varchar (2),
    cep varchar(8),
    codigo_pais varchar(4),
    pais nvarchar (30),
    fone nvarchar(12),
    inscricao_estadual varchar (12)
)"""
script.append(table_emitente)  #sempre adicionando ao fim da lista o script
table_dest = """CREATE TABLE Destinatarios (
    id int AUTO_INCREMENT PRIMARY KEY,
    cnpj varchar(14),
    nome_empresa nvarchar (50),
    logradouro nvarchar (100),
    numero nvarchar(8),
    complemento nvarchar (50),
    bairro nvarchar (30),
    cod_municipio varchar(7),
    municipio nvarchar (20),
    estado varchar (2),
    cep varchar(8),
    codigo_pais varchar(4),
    pais nvarchar (30),
    fone nvarchar(12),
    inscricao_estadual varchar (12)
)"""
script.append(table_dest)
table_transp  = """CREATE TABLE Transportadoras (
    id int AUTO_INCREMENT PRIMARY KEY,
    modo_frete int,
    cnpj varchar(14),
    nome_empresa nvarchar (50),
    inscricao_estadual varchar (12),
    endereco nvarchar(200),
    municipio nvarchar (20),
    estado varchar (2)
)"""
script.append(table_transp)
table_nf = """CREATE TABLE Notas_Fiscais(
    id int AUTO_INCREMENT PRIMARY KEY,
    cod_UF int,
    cod_NF nvarchar(12),
    Operaçao nvarchar (20),
    indicador_pgto int,
    modelo int,
    serie int,
    numero_nota int,
    data_Emissao datetime,
    data_saida_entrega datetime,
    tipo_nf int,
    cod_municipio int,
    formato_emiss int,
    tipo_emissao int,
    digito_verificador int,
    ver_ambiente int,
    finalidade_emissao int,
    processo_emissa int,
    verProc varchar (50),
    base_calculo float,
    valor_ICMS float,
    valor_base_ST float,
    valor_ST float,
    valor_prods float,
    valor_Frete float,
    valor_Seg float,
    valor_Desc float,
    valor_II  float,
    valor_IPI float,
    valor_PIS float,
    valor_COFINS float,
    valor_Outro float,
    valor_total_NF float,
    emissor int not null,
       CONSTRAINT fk_emissor
    FOREIGN KEY (emissor)
        REFERENCES Emitentes(id),
    destinatario int not null,
    CONSTRAINT fk_destinatario
    FOREIGN KEY (destinatario)
        REFERENCES Destinatarios(id),
    transportadora int not null,
    CONSTRAINT fk_transportadora
    FOREIGN KEY (transportadora)
        REFERENCES Transportadoras(id)
)"""
script.append(table_nf)
table_veiculos = """CREATE TABLE Veiculos(
    placa varchar (7),
    UF varchar (2),
    RNTC int,
    id_transp int not null,
    CONSTRAINT fk_id_transp
    FOREIGN KEY (id_transp)
        REFERENCES Transportadoras(id)
)"""
script.append(table_veiculos)
table_reboque = """CREATE TABLE Reboques(
    id int AUTO_INCREMENT PRIMARY KEY, 
    placa varchar (7),
    UF varchar (2),
    RNTC int,
    id_transp int not null,
    CONSTRAINT fk_id_transp_r
    FOREIGN KEY (id_transp)
        REFERENCES Transportadoras(id)
)"""
script.append(table_reboque)
table_vol = """CREATE TABLE Volume_Reboques(
    volume int,
    especie_volume nvarchar(20),
    marca nvarchar(20),
    num_volumes int,
    peso_liq float,
    peso_brut float,
    lacre varchar(11),
    id_reboque int not null,
    CONSTRAINT fk_reboque_vr
    FOREIGN KEY (id_reboque)
        REFERENCES Reboques(id),
    id_nota int not null,
    CONSTRAINT fk_nota_vr
    FOREIGN KEY (id_nota)
        REFERENCES Notas_Fiscais(id)
)"""
script.append(table_vol)
table_prod = """CREATE TABLE Produtos(
    id int AUTO_INCREMENT PRIMARY KEY,
    cod_Prod varchar(5),
    nome_prod nvarchar (50),
    Codigo_Fiscal int,
    unidade_comercial nvarchar (15),
    valor_unitario int,
    id_nota int not null,
    CONSTRAINT fk_id_nota
    FOREIGN KEY (id_nota)
        REFERENCES Notas_Fiscais(id)
)
"""
script.append(table_prod)
table_nitens = """CREATE TABLE ITENS_NOTA(
    nitem varchar(5),
    qtd_prod float,
    valor_unitario int,
    valor_produto float,
    unidade_medida_trib nvarchar (15),
    quantidade_trib float,
    valor_unitario_trib int,
    ICMS_origem int,
    ICMS_cst int,
    ICMS_modbc int,
    ICMS_vBC float,
    ICMS_pIMCS float,
    ICMS_vICMS float,
    PIS_cst int,
    PIS_vBC float,
    PIS_pPis float,
    PIS_vPis float,
    COFINS_cst int,
    COFINS_vBc float,
    COFINS_pCofins float,
    COFINS_vCofins float,
    id_nota int not null,
    CONSTRAINT fk_nota
    FOREIGN KEY (id_nota)
        REFERENCES Notas_Fiscais(id),
    id_produto int not null,
    CONSTRAINT fk_prod
    FOREIGN KEY (id_produto)
        REFERENCES Produtos(id)
)"""
script.append(table_nitens)
table_entrega = """CREATE TABLE Entregas(
    cnpj varchar(14),
    logradouro nvarchar (100),
    numero nvarchar(8),
    complemento nvarchar (50),
    bairro nvarchar (30),
    cod_municipio varchar(7),
    municipio nvarchar (20),
    estado varchar (2),
    id_transportadora int not null,
    CONSTRAINT fk_transportadora_entrega
    FOREIGN KEY (ID_transportadora)
        REFERENCES Transportadoras(id),
    id_nota int not null,
    CONSTRAINT fk_id_nota_entrega
    FOREIGN KEY (id_nota)
        REFERENCES Notas_Fiscais(id)
)"""
script.append(table_entrega)
table_entrega = """CREATE TABLE Retiradas(
    cnpj varchar(14),
    logradouro nvarchar (100),
    numero nvarchar(8),
    complemento nvarchar (50),
    bairro nvarchar (30),
    cod_municipio varchar(7),
    municipio nvarchar (20),
    estado varchar (2),
    id_transportadora int not null,
    CONSTRAINT fk_transportadora_retirada
    FOREIGN KEY (ID_transportadora)
        REFERENCES Transportadoras(id),
    id_nota int not null,
    CONSTRAINT fk_id_nota_retirada
    FOREIGN KEY (id_nota)
        REFERENCES Notas_Fiscais(id)
)"""
script.append(table_entrega)
```
